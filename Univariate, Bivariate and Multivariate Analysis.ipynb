```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
```

```python
df.head()
```

### Univariate analysis

```python
df.loc[df['species']=='setosa'] ## locating all the setosa points
```

```python
df_setosa = df.loc[df['species']=='setosa'] 
```

```python
df_virginica = df.loc[df['species']=='virginica'] 
df_versicolor = df.loc[df['species']=='versicolor'] 
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']))
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']))
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length'])) # o parameter makes the points bigger
plt.xlabel('Sepal Length')
plt.show()

# in the univariate we're keeping all the y-axis as zeroes
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o') # o parameter makes the points bigger
plt.xlabel('Sepal Length')
plt.show()

# in the univariate we're keeping all the y-axis as zeroes
```

### Z-Score Statistics

When converting a guassian normal distribution to a standard normal distribution with mean 0 and s.d = 1, we use Z-score

Formula: 

Z-score = x(i) - mu / sigma

For example, consider: {1,2,3,4,5} for this mean is 3 and S.D is 1. Apply the Z-score formula

for 3, z-score = 3-3/1 --> 0, and 4 will be 1 and 2 will be -1
